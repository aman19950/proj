## Functions
- it is a piece code of block

```python
# function with no argument with no return
def add():
    print("hello")

add()
add()
add()
```

```python
# function with argument with no return
def add(x,y): #formal arguments
    print("add--",x+y)
    print("multi-",x*y)
    
add(4,5) #actual arguments
add(5,5)
```

```python
# function with no argument with return

def xyz():
    z = 3 + 5
    y = 4  * 6
    return z,y
c,d = xyz()
print(c)
print(d)
```

```python
#function with argument with return

def xyz(x,y):
    z =  x+y
    return z

c = xyz(3,15)
print(c)
```

```python
# *arg --------- **kwargs
def xyz(y,*x):
    print(y)
    print(x)

xyz(1,2,3,"hello",7,8,9)
```

```python
def xyz(*x,**y):
    print(x)
    print(y)

xyz(1,2,3,a=100,b=200,c=300)
```

```python
x = 10                   # global variable
y = 20                    # global variable
def xyz():
    global x
    x = 100
    y = 200               # local variable
    z = x + y             # z is a local variable
    return z

print(x) 
c = xyz()
print(x)
```

```python
def xyz(x,y):
    z = x + y
    print(z)
    def abc():
        nonlocal z
        z = 6 + 6
        print(z)
    abc()
    print(z)
    
xyz(4,5)
```

## Recurrsion

```python
# Function calling itself

def xyz():
    print("hello")
    xyz()

xyz()
```

```python
# wap a program to print factorial of a number using recursion
x = int(input("enter a number = "))

def fact(n):
    if n == 0 or n == 1:
        return 1
    else:
        return n * fact(n-1)
    

print(fact(x))
```

```python
# wap to print fibbonacci series using recursion
x = int(input("enter a number = "))

def fib(n):
    if n<= 1:
        return n
    else:
        return fib(n-1)+fib(n-2)
    
if x < 0:
    print("series doesn't exists")
else:
    for i in range(x):
        print(fib(i))
```
